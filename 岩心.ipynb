```python
import pandas as pd

# 初始化列表
label_y = []
image_name = []

# 根据图片序号分配类别
for i in range(1, 321):  # 假设有320张图片，序号从1开始
    image_name.append(f'{i}.jpg')
    if i <= 80:  # 第1-80张图片分配给类别0
        label_y.append(0)
    elif i <= 160:  # 第81-160张图片分配给类别1
        label_y.append(1)
    elif i <= 240:  # 第161-240张图片分配给类别2
        label_y.append(2)
    else:  # 剩余的图片分配给类别3
        label_y.append(3)

# 创建DataFrame
dataframe = pd.DataFrame({'label_y': label_y, 'image_name': image_name})

# 将DataFrame存储为CSV文件
dataframe.to_csv(r"D:\2067455485\FileRecv\CNN dataset\train_label.csv", index=False, sep=',')
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from torch.utils.data import Dataset, DataLoader
import tensorflow as tf
from tensorflow.keras.layers import Conv2D, BatchNormalization, Activation, MaxPooling2D, Dropout, Flatten, Dense, Input
from tensorflow.keras.models import Sequential
from tensorflow.keras.utils import to_categorical
import datetime

# 设置matplotlib支持中文字体
plt.rcParams['font.sans-serif'] = ['SimHei']  # 指定默认字体
plt.rcParams['axes.unicode_minus'] = False  # 解决负号'-'显示为方块的问题

# 加载数据集
class Custom_Dataset(Dataset):
    def __init__(self, img_dir, img_label_dir, transform=None):
        super().__init__()
        self.img_dir = img_dir
        self.img_labels = pd.read_csv(img_label_dir)
        self.transform = transform

    def __len__(self):
        return len(self.img_labels)

    def __getitem__(self, index):
        img_name = self.img_labels.iloc[index, 1]
        img_path = os.path.join(self.img_dir, img_name)
        
        if not os.path.exists(img_path):
            raise FileNotFoundError(f"File not found: {img_path}")
        
        image = plt.imread(img_path)
        
        if image is None:
            raise ValueError(f"Failed to load image: {img_path}")
        
        label = self.img_labels.iloc[index, 0]
        return image, label

# 标签指示含义
label_dic = {0: '膏岩', 1: '灰岩', 2: '灰质膏岩', 3: '膏质灰岩'}

# 图像集及标签路径
label_path = r"D:\2067455485\FileRecv\CNN dataset\train_label.csv"
img_root_path = r"D:\2067455485\FileRecv\CNN dataset\train_revise"
test_image_path = r"D:\2067455485\FileRecv\CNN dataset\test_revise"
test_label_path = r"D:\2067455485\FileRecv\CNN dataset\test_label.csv"

# 实例化类
dataset = Custom_Dataset(img_root_path, label_path)
dataset_test = Custom_Dataset(test_image_path, test_label_path)

# 将图像转变为网络可用的数据类型
try:
    x_train_list, y_train_list = [], []
    for imgs, labels in DataLoader(dataset, batch_size=320, shuffle=True):
        for img, label in zip(imgs, labels):
            x_train_list.append(img.numpy())
            y_train_list.append(label.item())
    x_train = np.array(x_train_list)
    y_train = np.array(y_train_list)
    
    x_test_list, y_test_list = [], []
    for imgs, labels in DataLoader(dataset_test, batch_size=40, shuffle=True):
        for img, label in zip(imgs, labels):
            x_test_list.append(img.numpy())
            y_test_list.append(label.item())
    x_test = np.array(x_test_list)
    y_test = np.array(y_test_list)
    
    x_train = x_train.astype('float32') / 255.0
    x_test = x_test.astype('float32') / 255.0
    y_train = to_categorical(y_train, num_classes=4)
    y_test = to_categorical(y_test, num_classes=4)
except Exception as e:
    print(f"Error creating training and testing sets: {e}")

# CNN模型
model = Sequential([
    Input(shape=(256, 256, 3)),
    Conv2D(32, (5, 5), activation='relu'),
    MaxPooling2D((4, 4), padding='same'),
    MaxPooling2D((2, 2), padding='same'),
    Dropout(0.2),
    Flatten(),
    BatchNormalization(),
    Dense(50, activation='relu'),
    Dense(4, activation='softmax')
])

# 模型参数展示、编译与训练
model.summary()
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
startdate = datetime.datetime.now()
history = model.fit(x_train, y_train, validation_data=(x_test, y_test), epochs=40, batch_size=5, verbose=2)
enndate = datetime.datetime.now()
print("训练用时:" + str(enndate - startdate))

# 模型损失值与精度画图展示
loss = history.history['loss']  # 训练集损失
val_loss = history.history['val_loss']  # 测试集损失
acc = history.history['accuracy']  # 训练集准确率
val_acc = history.history['val_accuracy']  # 测试集准确率

plt.figure(figsize=(8, 4))
plt.subplot(121)
plt.plot(loss, color='b', label='训练损失')
plt.plot(val_loss, color='r', label='测试损失')
plt.ylabel('损失')
plt.legend()

plt.subplot(122)
plt.plot(acc, color='b', label='训练准确率')
plt.plot(val_acc, color='r', label='测试准确率')
plt.ylabel('准确率')
plt.legend()
plt.title('训练与测试结果')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.utils import to_categorical
import datetime

# 设置matplotlib支持中文字体
plt.rcParams['font.sans-serif'] = ['SimHei']  # 指定默认字体
plt.rcParams['axes.unicode_minus'] = False  # 解决负号'-'显示为方块的问题

# 假设x_train, y_train, x_test, y_test已经准备好，并且y_train和y_test已经是one-hot编码

# 将图像数据扁平化
x_train_flat = x_train.reshape(x_train.shape[0], -1)
x_test_flat = x_test.reshape(x_test.shape[0], -1)

# 全连接神经网络模型
model = Sequential([
    Dense(64, activation='relu', input_shape=(196608,)),  # 输入层，输入形状为(196608,)
    Dense(64, activation='relu'),  # 隐藏层
    Dense(4, activation='softmax')  # 输出层，4个神经元对应4个类别
])

# 模型参数展示、编译与训练
model.summary()
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
startdate = datetime.datetime.now()
history = model.fit(x_train_flat, y_train, validation_data=(x_test_flat, y_test), epochs=40, batch_size=5, verbose=2)
enddate = datetime.datetime.now()
print("训练用时:" + str(enddate - startdate))

# 模型损失值与精度画图展示
loss = history.history['loss']  # 训练集损失
val_loss = history.history['val_loss']  # 测试集损失
acc = history.history['accuracy']  # 训练集准确率
val_acc = history.history['val_accuracy']  # 测试集准确率

plt.figure(figsize=(8, 4))
plt.subplot(121)
plt.plot(loss, color='b', label='训练损失')
plt.plot(val_loss, color='r', label='测试损失')
plt.ylabel('损失')
plt.legend()

plt.subplot(122)
plt.plot(acc, color='b', label='训练准确率')
plt.plot(val_acc, color='r', label='测试准确率')
plt.ylabel('准确率')
plt.legend()
plt.title('训练与测试结果')
plt.show()
```

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import os

# 绘制箱线图
def plot_boxplot(labels, image_path, title, num_samples=50):
    data = []
    labels_unique = labels['label_y'].unique()
    for label in labels_unique:
        label_data = []
        for index in labels[labels['label_y'] == label].index[:num_samples]:
            image_name = labels.iloc[index, 1]  # 假设第二列是图片名称
            image_full_path = os.path.join(image_path, image_name)
            if os.path.isfile(image_full_path):
                img = Image.open(image_full_path).convert('L')  # 转换为灰度图
                img_array = np.array(img)
                label_data.append(img_array.flatten())  # 将图片数据展平
        if label_data:
            data.append(np.concatenate(label_data))
    if data:
        plt.boxplot(data, labels=labels_unique)
        plt.title(title)
        plt.xlabel('Label')
        plt.ylabel('Pixel Value')
        plt.show()

# 假设图片路径
test_image_path = r'D:\2067455485\FileRecv\CNN dataset\test_revise'
train_image_path = r'D:\2067455485\FileRecv\CNN dataset\train_revise'

# 读取测试集和训练集标签
test_labels = pd.read_csv(test_label_path)
train_labels = pd.read_csv(train_label_path)

# 绘制测试集箱线图
plot_boxplot(test_labels, test_image_path, 'Pixel Value Distribution - Test Set')

# 绘制训练集箱线图
plot_boxplot(train_labels, train_image_path, 'Pixel Value Distribution - Train Set')
```
